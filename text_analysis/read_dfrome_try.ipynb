## テキスト取り出し

```python
import pandas as pd
```

```python
# ノード情報

nodes = d["data"]["items"]["artifacts"]
nodes[1]
```

```python
len(nodes)
```

```python
# テキストすべて

sent = [nodes[i]['fT'] + " " +nodes[i]['sT'] for i in range(len(nodes))]
sent
```

```python
# .txtファイルに保存

with open('%s.txt' %file, 'w') as f:
    for line in sent:
        f.write(line)
        f.write('\n')
```

```python
lst = [[nodes[i]['id'], nodes[i]['fT'], nodes[i]['sT'], nodes[i]['pos']['x'], nodes[i]['pos']['y']] for i in range(len(nodes))]
lst[:2]
df = pd.DataFrame(lst, columns =['id', 'fT', 'sT', 'x', 'y']) 
print(df.head())
print(df.describe())
```

```python
# リスクシナリオ・マイノリティにとってうれしくないシナリオ
df_risk = df[df['x'] > 900]
df_risk.describe()
```

```python
risk = (df_risk['fT'] + ' ' + df_risk['sT']).to_list()
risk
```

```python
# ベストシナリオ・マイノリティもマジョリティもハッピーなシナリオ

df_best = df[df['x'] < 900]
df_best.describe()
df_best.head()
```

```python
best = (df_best['fT'] + ' ' + df_best['sT']).to_list()
best
```

```python
# form a dataframe of risk and best texts with a label

df1 = pd.DataFrame(df_risk['fT'] + ' ' + df_risk['sT'], columns=['text'])
df1['scenario'] = 'risk'
df2 = pd.DataFrame(df_best['fT'] + ' ' + df_best['sT'], columns=['text'])
df2['scenario'] = 'best'
df_t = pd.concat([df1, df2]).reset_index(drop=True)
# df_t.drop(columns=['index'])
df_t.head()
# df_t.describe()
```
