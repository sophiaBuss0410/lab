```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.1, 5, 50)
y1 = np.log(x)
y2 = -np.log(x)
flat = np.zeros_like(x)

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, flat)
plt.legend(["logx","- logx"])
plt.show()
```
