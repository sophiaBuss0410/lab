```python
from matplotlib import pyplot as plt
import networkx as nx

def show_graph(in_list):
  plt.figure(figsize=(4,6))
  G = nx.Graph()
  rels = in_list
  G.add_edges_from(rels)
  nx.draw_networkx(G, node_color='moccasin')
  plt.savefig('human_network.png')
  plt.show()
```
